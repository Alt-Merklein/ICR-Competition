```python
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_csv("../data/train.csv", index_col="Id")
```

```python
columns = ["AB","AF","AH","AM","AR","AX","AY","AZ","Class"]
data = data[columns]

#Filtering outliers
z_scores = stats.zscore(data)
threshold = 3
data_filtered = data[(z_scores < threshold).all(axis=1)]
data.head()
```
