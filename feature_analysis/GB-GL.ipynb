```python
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_csv("../data/train.csv", index_col="Id")
```

```python
columns = ["GB","GE","GF","GH","GI","GL","Class"]
data = data[columns]

#filling a few NaN entries
data["GL"] = data["GL"].fillna(data["GL"].mean())


#Filtering outliers
z_scores = stats.zscore(data) #EJ is categorical
threshold = 3
data_filtered = data[(z_scores.abs() < threshold).all(axis=1)]
data_filtered.head()
```
