```python
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_csv("../data/train.csv", index_col="Id")
```

```python
columns = ["FC","FD ","FE","FI","FL","FR","FS","Class"]
data = data[columns]

#filling a few NaN entries
data["FC"] = data["FC"].fillna(data["FC"].mean())
data["FS"] = data["FS"].fillna(data["FS"].mean())
data["FL"] = data["FL"].fillna(data["FL"].mean())

#Filtering outliers
z_scores = stats.zscore(data)
threshold = 3
data_filtered = data[(z_scores.abs() < threshold).all(axis=1)]
data_filtered.head()
```
