```python
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_csv("../data/train.csv", index_col="Id")
```

```python
columns = ["BC","BD ","BN","BP","BQ","BR","BZ","Class"]
data = data[columns]
data["BQ"] = data["BQ"].fillna(data["BQ"].mean()) #Since the column BQ has missing entries
#Filtering outliers
z_scores = stats.zscore(data)
threshold = 3
data_filtered = data[(z_scores < threshold).all(axis=1)]
data_filtered.head()
```
