```python
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_csv("../data/train.csv", index_col="Id")
```

```python
columns = ["CB","CC","CD ","CF","CH","CL","CR","CS","CU","CW ","Class"]
data = data[columns]

#filling a few NaN entries
data["CB"] = data["CB"].fillna(data["CB"].mean())
data["CC"] = data["CC"].fillna(data["CC"].mean())

#Filtering outliers
z_scores = stats.zscore(data)
threshold = 3
data_filtered = data[(abs(z_scores) < threshold).all(axis=1)]
data_filtered.head()
```
