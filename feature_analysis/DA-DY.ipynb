```python
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_csv("../data/train.csv", index_col="Id")
```

```python
columns = ["DA","DE","DF","DH","DI","DL","DN","DU","DV","DY","Class"]
data = data[columns]
#filling a few NaN entries
data["DU"] = data["DU"].fillna(data["DU"].mean())

#Filtering outliers
z_scores = stats.zscore(data)
threshold = 3
data_filtered = data[(z_scores.abs() < threshold).all(axis=1)]
data_filtered.head()
```
